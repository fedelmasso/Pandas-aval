```python
# 1) Apresente em tela (output) toda a base de dados.

import pandas as pd

filmes =['Um Drink no Inferno', 'Cães de Aluguel', 'Django Livre', 'Bastardos Inglórios', 'Era uma Vez em Hollywood', 'Kill Bill']
genero =["Terror/Ação", "Crime/Thriller", "Faroeste/Ação", "Guerra/Aventura", "Comédia/Faroeste", "Ação/Thriller"]
duracao =['1h48m', '1h39m', '2h45m', '2h33m', '2h40m', '1h51m']
ano_lancamento =[1996, 1992, 2012, 2009, 2019, 2003]
streaming_disponiveis =["Amazon Prime", "Telecine", "Netflix", "Star+", "Amazon Prime", "Telecine"]
classificacao_etaria =[18, 18, 16, 18, 18, 16]

info_filme ={
    'Filmes': filmes,
    'Gênero': genero,
    'Duração': duracao,
    'Lançamento': ano_lancamento,
    'Streaming': streaming_disponiveis,
    'Classificação Indicativa ': classificacao_etaria
}
print("\n")

#criando o dataframe a partir do dicionario
filmes_df = pd.DataFrame(info_filme)
display(filmes_df)
```

```python
# 2) Apresente o tamanho do seu dataframe (quantas colunas x linhas).
print (filmes_df.shape)
```

```python
# 3) Acesse a linha (x) e apresente em tela todas as características do item.
filmes_df.loc()[0]
```

```python
# 4) Verifique se o dataframe está vazio.
print (filmes_df.empty)
```

```python
# 5) Apresente em tela os 5 primeiros registros da base de dados.
primeiros_registros = filmes_df.head(5)
display(primeiros_registros)
```

```python
# 6) Exclua um item (linha) de sua base de dados.
filmes_df = filmes_df.drop(0)
display(filmes_df)
```

```python
# 7) Adicione um item (linha) na sua base de dados
novo_filme = {
    'Filmes': 'Pulp Fiction',
    'Gênero': 'Crime/Drama',
    'Duração': '2h34m',
    'Lançamento': 1994,
    'Streaming': 'Netflix',
    'Classificação Indicativa ': 18
}

filmes_df.loc[len(filmes_df.index+1)] = novo_filme
display(filmes_df)
```

```python
# 8) Transponha a coluna para a linha em sua base de dados.
transposto = filmes_df.transpose()
print(transposto)
```

```python
# 9) Apresente em tela somente a 1ª e a 2ª coluna (rótulo) da base de dados.
primeira_segunda_coluna = filmes_df.iloc[:, :2]  #Usa o método iloc para selecionar partes do DataFrame com base na posição. Aqui, [:, :2] significa "todas as linhas" (:) e "as duas primeiras colunas" (:2).
display(primeira_segunda_coluna) #Exibe o resultado da seleção das duas primeiras colunas do DataFrame.
```
